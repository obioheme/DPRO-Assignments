# Creating Dummy Data

```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
```

### Problem 1: Creating Random Numbers

```python
mean = [-3, 0]
cov = [[1.0, 0.8], [0.8, 1.0]]

np.random.seed(0)
array = np.random.multivariate_normal(mean, cov, (500))
#array = np.random.random((500, 2))
array
```

### Problem 2: Visualization by Scatter Plot

```python
plt.scatter(array[:, 0], array[:, 1])
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Scatter Plot")
plt.show()
```

### Problem 3: Visualization by Histogram

```python
plt.hist(array[:, 0])
plt.xlabel("x1")
plt.ylabel("Frequency")
plt.title("Histogram of x1")
plt.xlim(right=2)
plt.show()

plt.hist(array[:, 1])
plt.xlabel("x2")
plt.ylabel("Frequency")
plt.title("Histogram of x2")
plt.xlim(left=-5)
plt.show()
```

### Problem 4: Addition of Data

```python
mean = [0, -3]
cov = [[1.0, 0.8], [0.8, 1.0]]

np.random.seed(0)
array2 = np.random.multivariate_normal(mean, cov, (500))
#array = np.random.random((500, 2))
array2

#Visualizing on a scatter plot
fig, figure = plt.subplots()
figure.scatter(array[:, 0], array[:, 1], label="0")
figure.scatter(array2[:, 0], array2[:, 1], color="r", label="1")
figure.set_xlabel("x1")
figure.set_ylabel("x2")
figure.set_title("Scatter Plot")
figure.legend(loc="upper right")
plt.show()
```

### Problem 5: Data Combination

```python
array3 = np.vstack([array, array2])
print(array3.shape)
```

### Problem 6: Labelling

```python
#Create a new column with zeros and ones
new_col = np.vstack([np.zeros((500, 1)), np.ones((500, 1))])
print(new_col.shape)

#Concatenate the new column with the array
array3 = np.concatenate([array3, new_col], axis =1)
array3
```
