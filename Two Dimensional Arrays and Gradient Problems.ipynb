# Two Dimensional Array and Gradient Problem

### Problem 1: Linear Function

```python
import numpy as np
```

```python
x = np.arange(-50, 50.1, 0.1)
y = (0.5*x) - 1
print(x)
print(y)
```

### Problem 2: Array Combination

```python
xy = np.c_[x, y]
print(xy)
print(xy.shape)
```

### Problem 3: Find the Gradient

```python
x_diff = x[:-1] - x[1:]
y_diff = y[:-1] - y[1:]

grad = y_diff/x_diff
grad.shape
```

### Problem 4: Draw a Graph

```python
import matplotlib.pyplot as plt

fig, ((axis1, axis2)) = plt.subplots(1, 2, figsize=(10, 5))
axis1.plot(xy[:, 0], xy[:, 1])
axis1.set_xlabel("x")
axis1.set_ylabel("y")
axis1.set_title("Plot of Y against X")

axis2.plot(xy[1:, 0], grad)
axis2.set_xlabel("x")
axis2.set_ylabel("Gradient")
axis2.set_title("Plot of Gradient against X")
```

### Problem 5: Python Functionalization

```python
def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    
    x = np.arange(x_range[0], x_range[1], x_range[2])
    y = function(x)
    xy = np.c_[x, y]
    
    x_diff = x[1:] - x[:-1]
    y_diff = y[1:] - y[:-1]
    grad = y_diff/x_diff
    
    
    return xy, grad

def function1(x):
    pass
    return y

xy, grad = compute_gradient(function1)
```

```python
def function2(x):
    y = x**2
    return y

xy2, grad2 = compute_gradient(function2)
```

```python
fig, ((axis3, axis4)) = plt.subplots(1, 2, figsize=(10, 5))
axis3.plot(xy2[:, 0], xy2[:, 1])
axis3.set_xlabel("x")
axis3.set_ylabel("y")
axis3.set_title("Plot of Y = x^2")

axis4.plot(xy2[1:, 0], grad2)
axis4.set_xlabel("x")
axis4.set_ylabel("Gradient")
axis4.set_title("Plot of Gradient against X")
```

```python
def function3(x):
    y = (2*x*x) + (2**x)
    return y

xy3, grad3 = compute_gradient(function3)
```

```python
fig, ((axis5, axis6)) = plt.subplots(1, 2, figsize=(10, 5))
axis5.plot(xy3[:, 0], xy3[:, 1])
axis5.set_xlabel("x")
axis5.set_ylabel("y")
axis5.set_title("Plot of Y = 2x^2 + 2^x")

axis6.plot(xy3[1:, 0], grad3)
axis6.set_xlabel("x")
axis6.set_ylabel("Gradient")
axis6.set_title("Plot of Gradient against X")
```

```python
def function4(x):
    ang = x ** 0.5
    y = np.sin(ang)
    return y

xy4, grad4 = compute_gradient(function4, x_range=(0, 50.1, 0.1))
```
