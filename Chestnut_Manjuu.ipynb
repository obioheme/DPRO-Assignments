# Chestnut Manjuu Problem

```python
def number_of_days():
    solar_system = 12 * (10**26) # size in cubic miles
    #convert cubic miles to cubic meters
    s_system = solar_system * 1609.34

    chestnut_size = 23 # size in grams
    #convert grams to cubic meters
    c_size = chestnut_size * 0.00000001
    time = 0
    num_chestnut = 1
    while c_size <= s_system:
        time +=5 #time in minutes
        num_chestnut *=2 #chestnut doubles every 5 minute. After 5 minute the total chestnuts also doubles again
        c_size *= 2
    #convert time in minutes to days since 1 day is equal to 1440 minutes
    days = time / 1440
    print(f"It takes {days :.2} days to fill the solar system with chestnut manjuu of size {chestnut_size}g.")

number_of_days()
```
