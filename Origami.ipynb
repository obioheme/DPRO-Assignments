# Using Exponentiation Arithmetic Operators

```python
import time
start = time.time()

t0 = 0.00008
num_of_folds = 43
folded_thickness = (t0 * (2**num_of_folds)) / 1000
print("Thickness {: .2} kilometers".format(folded_thickness))

elasped_time = time.time() - start
print("time: {}[s]".format(elasped_time))
```

# Using For Statement

```python
import time
start = time.time()

folded_thickness = 1
for folds in range(1, 44):
    thickness = 2
    folded_thickness *= thickness
folded_thickness = (folded_thickness * 0.00008) / 1000     
print("Thickness {: .2} kilometers".format(folded_thickness))    

elasped_time = time.time() - start
print("time: {}[s]".format(elasped_time))
```

```python
%%timeit

folded_thickness = 1
for folds in range(1, 44):
    thickness = 2
    folded_thickness *= thickness
folded_thickness = (folded_thickness * 0.00008) / 1000     
print("Thickness {: .2} kilometers".format(folded_thickness)) 
```

```python
thickness_list = []
folded_thickness = 0.00008
thickness_list.append(folded_thickness)

for folds in range(1, 44):
    thickness = 2
    folded_thickness *= thickness
    thickness_list.append(folded_thickness)

print(len(thickness_list))
```

```python
import matplotlib.pyplot as plt

plt.title("Thickness of Folded Paper")
plt.xlabel("Number of Folds")
plt.ylabel("Thickness [m]")
plt.plot(thickness_list, color ="y", linestyle = ":", linewidth = 4)
plt.show()
```
